# Description

This is a Indian data set for the period of 2001-2012.And even 2013 data set seperately. Data is related to Crime head-wise persons arrested under crime against children i.e., crime is done and conviceted by criminals. this data is represents all the states and Union territories.

```python
import os
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
crime= pd.read_csv('Crime head-wise persons arrested under crime against children during 2001-2012.csv', sep=',')
```

```python
crime.columns
```

```python
crime.info()
```

```python
crime.isnull().sum()
```

```python
crime.columns=crime.columns.str.strip()
crime.columns=crime.columns.str.lower()
crime.columns= crime.columns.str.replace(' ','_')
crime.columns= crime.columns.str.replace('/','_')
```

```python
crime1= crime
```

```python
crime1=crime.drop(columns=['total'])
```

```python
crime1.columns
```

```python
crime_melt = pd.melt(crime1, # data to reshape
                            id_vars = ['state_ut', 'crime_head'], # identifier variable
                            var_name = 'year') # t
```

```python
crime_melt
```

```python
crime_melt['crime_head'].describe()
```

```python
crime_melt['crime_head'].value_counts()
```

```python
crime_melt.loc[crime_melt.crime_head.str.contains('ABETMENT OF SUICIDE|KIDNAPPING and ABDUCTION OF CHILDREN|MURDER OF CHILDREN'),'crime_classification']= 'Abduction And Abetment Against Children'
```

```python
crime_melt.loc[crime_melt.crime_head.str.contains('FOETICIDE|INFANTICIDE'),'crime_classification']= 'Abortion Or Infanticide'
```

```python
crime_melt.loc[crime_melt.crime_head.str.contains('BUYING OF GIRLS FOR PROSTITUTION|SELLING OF GIRLS FOR PROSTITUTION'),'crime_classification']= 'Prostitution'
```

```python
crime_melt.loc[crime_melt.crime_head.str.contains('RAPE OF CHILDREN|PROCURATION OF MINOR GILRS|PROHIBITION OF CHILD MARRIAGE ACT|EXPOSURE AND ABANDONMENT'),'crime_classification']= 'Crime Against Minor Girls'
```

```python
crime_melt.loc[crime_melt.crime_head.str.contains('OTHER CRIMES AGAINST CHILDREN'),'crime_classification']= 'OTHER CRIMES AGAINST CHILDREN'
```

```python
crime_melt['crime_classification'].value_counts()
```

```python
crime_melt['state_ut'].describe()
```

```python
crime_melt
```

```python
crime_melt.pivot_table(index='state_ut', columns='crime_classification', values='value')
```

```python
crime_melt.pivot_table(index='state_ut', columns='crime_classification', values='value', aggfunc='sum')
```

```python
crime_melt.pivot_table(index='state_ut', columns='crime_classification',aggfunc={'value':sum, 'year':max})
```

```python
sns.set()  # use Seaborn style
crime_melt.pivot_table(index='state_ut', columns='crime_classification', values='value', aggfunc='sum').plot(figsize=(12,6))
plt.ylabel('Total entries');
```

```python
crime_melt.loc[crime_melt.state_ut.str.contains('CHANDIGARH|MADHYA PRADESH'),'crime_state']= 'Central'
```

```python
crime_melt.loc[crime_melt.state_ut.str.contains('BIHAR|JHARKHAND|ODISHA|WEST BENGAL'),'crime_state']= 'Eastern'
```

```python
crime_melt.loc[crime_melt.state_ut.str.contains('ARUNACHAL PRADESH|ASSAM|MANIPUR|MEGHALAYA|MIZORAM|NAGALAND|SIKKIM|TRIPURA'),'crime_state']= 'North Eastern'
```

```python
crime_melt.loc[crime_melt.state_ut.str.contains('HIMACHAL PRADESH|JAMMU & KASHMIR|PUNJAB|RAJASTHAN|UTTAR PRADESH|UTTARAKHAND'),'crime_state']= 'Northern'
```

```python
crime_melt.loc[crime_melt.state_ut.str.contains('ANDHRA PRADESH|KARNATAKA|KERALA|TAMIL NADU'),'crime_state']= 'Southern'
```

```python
crime_melt.loc[crime_melt.state_ut.str.contaains('GOA|GUJARAT|MAHARASHTRA'),'crime_state']= 'Western'
```

```python
crime_melt['crime_state'].value_counts()
```

```python
crime_melt.pivot_table(index='crime_state', columns='crime_classification', values='value', aggfunc='mean')
```

```python
crime_melt.pivot_table(index='crime_state', columns='crime_classification', values='value', aggfunc='sum')
```

```python
sns.set()  # use Seaborn style
crime_melt.pivot_table(index='crime_state', columns='crime_classification', values='value', aggfunc='sum').plot(figsize=(12,6))
plt.ylabel('Total entries');
```

```python
sns.set()  # use Seaborn style
crime_melt.pivot_table(index='year', columns='crime_state', values='value', aggfunc='sum').plot(figsize=(12,6))
plt.ylabel('Total entries');
```
